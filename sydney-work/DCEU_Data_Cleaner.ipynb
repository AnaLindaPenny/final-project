# Marvel vs. DC - IMDB & Rotten Tomatoes Data Cleaner

Import tools needed

```python
import os
from bs4 import BeautifulSoup as bs
import requests
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

Import CSV

```python
csvfile = "marveldc.csv"
dc_data = pd.read_csv(csvfile, encoding='utf-8')
dc_data.head()
```

Isolate Marvel Movies

```python
dc_only = dc_data.loc[dc_data['Entity'] == 'DC', :]
dc_only.head()
```

Isolate Desired Columns

```python
dc_csv = dc_only[['Title', 'Year', 'Runtime', 'Director','IMDB_Rating', 'IMDB_Votes', 'IMDB_Gross', 'Tomato_Meter', 'Tom_Aud_Score']]
dc_csv.head()
```

Export to CSV

```python
dc_csv.to_csv("tableau_dc.csv", index=False, header=True)
```
