# Marvel vs. DC - IMDB & Rotten Tomatoes Data Cleaner

Import tools needed

```python
import os
from bs4 import BeautifulSoup as bs
import requests
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

Import CSV

```python
csvfile = "marveldc.csv"
marvel_data = pd.read_csv(csvfile, encoding='utf-8')
marvel_data.head()
```

Isolate Marvel Movies

```python
marvel_only = marvel_data.loc[marvel_data['Entity'] == 'MARVEL', :]
marvel_only.head()
```

Isolate Desired Columns

```python
marvel_csv = marvel_only[['Title', 'Year', 'Runtime', 'Director','IMDB_Rating', 'IMDB_Votes', 'IMDB_Gross', 'Tomato_Meter', 'Tom_Aud_Score']]
marvel_csv.head()
```

Export to CSV

```python
marvel_csv.to_csv("tableau_marvel.csv", index=False, header=True)
```
